```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[0,1,2,3,4,5]
lamb=2
```

```python
poisson_pd=poisson.pmf(x,lamb)
```

```python
fig,ax=plt.subplots(1,1,figsize=(10,10))
ax.plot(x,poisson_pd,label="Poisson_Pmf",color="purple",marker="H",lw=6,markersize=12)
plt.title("Poisson Distribution\n",fontsize=15,color="darkred")
plt.xlabel("Number of Resturants")
plt.ylabel("Probability")
ax.vlines(x,0,poisson_pd,color="darkorange",lw=5)
plt.legend(loc="best")
plt.show()
```
